# Generation of random variables

## Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Linear Congruential

```python
def runif(x, a, c, m, n):
    '''
    x: seed
    a: multiplicador
    c: incremento
    m: modulo
    n: quantidade de vari√°veis

    the best parameters: a = 48271, c = 0 if multiplicative or c = 1 if aditive , m = 2147483647.
    '''
    x = x
    x = (a * x + c) % m
    l_xp = []
    for i in range(n):
        x = (a * x + c) % m
        l_xp.append(x / m)
    return l_xp
```

```python
u = runif(27, 48271, 1, 2147483647, 1000000)
```

```python
# Plot do histograma
plt.hist(u, bins=30, color='lightgreen', edgecolor='black')
plt.title('Histograma dos 1000000 Valores Gerados pela Fun√ß√£o runif', fontsize=14)
plt.xlabel('Valores Uniformes (U)', fontsize=12)
plt.ylabel('Frequ√™ncia', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Exponencial

```python
def exponen(U, lambda__):
  U = np.array(U)
  return -1/lambda__ * np.log(U)
```

```python
e = exponen(runif(27, 48271, 1, 2147483647, 1000000), 0.5)
e[:5]
```

```python
# Plot do histograma
plt.hist(e, bins=30, color='red', edgecolor='black')
plt.title('Histograma dos 1000000 Valores Gerados pela Fun√ß√£o Inversa Exponencial', fontsize=14)
plt.xlabel('Valores Exponenciais', fontsize=12)
plt.ylabel('Frequ√™ncia', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Geometric distribution

```python
def geometric(U, p):
  U = np.array(U)
  return np.ceil(np.log(1 - U) / np.log(1-p))
```

```python
g = geometric(runif(27, 48271, 1, 2147483647, 1000000), 0.5)
g[:5]
```

```python
# Plot do histograma
plt.hist(g, bins=100, color='purple', edgecolor='black')
plt.title('Histograma dos 1000000 Valores Gerados pela Fun√ß√£o Inversa Geometrica', fontsize=14)
plt.xlabel('Valores Geometrica', fontsize=12)
plt.ylabel('Frequ√™ncia', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
